```python
import warnings
warnings.simplefilter('ignore')

import pandas as pd #imports pandas for later data manipulation
import matplotlib.pyplot as plt

df = pd.read_csv('s3://ia340-2025-fall-86749173517/DC_Public_Employee_Salary.csv') #reads the file into jupiter lab
df.head() #shows the first five rows of data
```

```python
df.shape #Displays the number of rows and columns within the dataset
```

```python
df.info() #Shows the column names the numebr of null values and the Dtype for each
```

```python
df.describe() #describes the statistical analysis of the data
```

```python
df = df.rename(columns={"JOBTITLE":"Job", "COMPRATE":"Pay", "HIREDATE_STRING":"Hire_Date"})  #Renames columns to easier titles for later 
df.head()
```

```python
df = df.dropna() #drops all null values in the dataset
df.isnull().sum() #Shows the total number of null values in each column
```

```python
df['Pay'] #Displays the different pay values
```

```python
df.sort_values(by='Hire_Date') #Sorts the data by the Hire date of the employee
```

```python
df[0:15] #Selects the first 15 rows of data to display
```

```python
df = df[(df['Pay']>=60000)] #Discards data with a pay value of less than $60,000 a year
df.describe()
```

```python
grouped_data = df.groupby('Hire_Date')['Pay'].mean()  #Groups the data by date and displays the average pay for the date hired
print(grouped_data)
```

```python
avg_salary_by_agency = (
    df.groupby("DESCRSHORT")["Pay"]
    .mean()
     .head(10)

)

avg_salary_by_agency.plot(kind="bar")
plt.title("Top 10 Agencies by Average Salary", fontsize=14)
plt.ylabel("Average Salary (COMPRATE)")
plt.xlabel("Agency (DESCRSHORT)")
plt.show()
```

```python
df['HIREDATE_STRING'] = pd.to_datetime(df['Hire_Date'])
df['HIRE_YEAR'] = df['HIREDATE_STRING'].dt.year

summary = df.groupby('HIRE_YEAR')['Pay'].agg(['min', 'mean', 'max']).reset_index()

plt.plot(summary['HIRE_YEAR'], summary['min'], label="Minimum")
plt.plot(summary['HIRE_YEAR'], summary['mean'], label="Mean")
plt.plot(summary['HIRE_YEAR'], summary['max'], label="Maximum")

plt.title("Min, Mean, and Max Salary by Hire Year (Line Plot)")
plt.xlabel("Hire Year")
plt.ylabel("Salary (Pay)")
plt.legend()
plt.show()
```
